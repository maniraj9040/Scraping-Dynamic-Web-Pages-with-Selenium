# locating elements in a webpage using selenium and python

```python
# importing necessary packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium .webdriver.chrome.options import Options
import time
```

```python
# adding options on chrome
chromeOptions = Options()
chromeOptions.add_argument('--kiosk')
```

# locating element by name/id

```python
driver = webdriver.Chrome(options = chromeOptions)
driver.get('https://www.pluralsight.com/')

#locate element by name
ele = driver.find_element_by_name('q')

#locate element by id
#ele = dirver.find_element_by_id('id-search-field')

time.sleep(1)     # waits until the page loads
ele.clear()       # clears everything from the search bar
ele.send_keys('python')   # types in the search bar
ele.send_keys(Keys.RETURN)   # "Enter" button is pressed
```

```python
driver.quit()
```

# Navigating a Web page with selenium and python

# 1.dealing with forms and dropdown menus

```python
# importing necessary packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
```

```python
# adding options on chrome
chromeOptions = Options()
chromeOptions.add_argument('--kiosk')
```

# 1.a. locating the elements in the form

```python
driver = webdriver.Chrome(options = chromeOptions)
driver.get('https://www.pluralsight.com/buy?requestId=158086ec-cae8-4076-ac03-6c42f1ac148e&priceOptionId=7842f8ee-b0f3-471c-a011-d7ab435e1593&legacyTrackingId=IND-M-PLUS&planId=Individual+plan&quantity=1&analyticsStep=1')

# web driver wait
ele = WebDriverWait(driver,10).until(
        EC.element_to_be_clickable((By.NAME,'firstName')))

#first name
firstname = driver.find_element_by_name('firstName')
firstname.clear()
firstname.send_keys('mani')

#last name

firstname = driver.find_element_by_name('lastName')
firstname.clear()
firstname.send_keys('raj')

# email
firstname = driver.find_element_by_name('email')
firstname.clear()
firstname.send_keys('mani123@gmail.com')

#confirmation email
firstname = driver.find_element_by_name('confirmationEmail')
firstname.clear()
firstname.send_keys('mani123@gmail.com')

```

# 1.b dropdown menu, checkbox, submit

```python
#country of residence

time.sleep(2)
country_element = driver.find_element_by_xpath("//select[@data-test='country']")
all_options = country_element.find_elements_by_tag_name('option')

for country in all_options:
    print(country.get_attribute('value'))
    if country.get_attribute('value') == 'IN':
        country.click()

#checkbox
time.sleep(1)
checkbox = driver.find_element_by_xpath("//div[@role = 'checkbox']")
checkbox.click()

#submit button
submit = driver.find_element_by_xpath("//button[@data-test = 'submitButton']")
time.sleep(2)
# submit.click()
```

# Extracting information from dynamic webpage

```python
# importing necessary packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
```

```python
# adding options on chrome
chromeOptions = Options()
chromeOptions.add_argument('--kiosk')
```

```python
# navigate to the premier leage site
driver = webdriver.Chrome(options = chromeOptions)
driver.get('https://www.premierleague.com')

#clicking on the players tab
player_ele = driver.find_element_by_link_text('Players').click()

#searching for wayne rooney
#web Driver Wait
element = WebDriverWait(driver,10).until(
            EC.element_to_be_clickable((By.ID,'search-input')))

search_ele = driver.find_element_by_id('search-input')
search_ele.send_keys('wayne rooney')
search_ele.send_keys(Keys.RETURN)

#clicking on wayne rooney

driver.implicitly_wait(2)
click_wayne = driver.find_element_by_xpath("//img[@data-player = 'p13017']").click()

```

There are two types of waits for webdriver 

1. Implicit wait  :- tells the web driver to wait for certain time period no matter what. 

2. Explicit wait  :- tells teh web driver to wait until a certain condition is satisfied.

# challenges while scraping


-> lack of bot access 

        -> websites are free to decide if they allow the bot.
        
        -> to see if the website is allowing bot access allways check robot.txt and terms of services 
        (ex: www.aaa.com/robot.txt)

-> CAPTCHA's

        -> implemented to keep spam away.
        
        -> Basic scripts tends to fail.
        
        -> Advancements have been made in both captcha's and workarounds.

-> Frequent Structure Changes

-> Getting blocked or baned

        -> unnatural high request from same IP - IP block
        
        -> Honeypot traps
        
        -> Bot Blocking services
        
-> Dynamic content

-> Questionable Data quality





# tips to increase Efficiency and best practices

-> Avoid loading images

        -> use chromeoptions to avoid loading images and selenium takes much time to load images

-> Employ disk cache

        -> can be implemented using chromeoptions
        
-> Close Drive with Quit



# Best Practices

-> check robot.txt file

-> don't hit webservers too frequently.

-> don't scrape during peak hours.

-> use headless browser.

-> avoid using same crawling pattern.

-> embrace transperency.
